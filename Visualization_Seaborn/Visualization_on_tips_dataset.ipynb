```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read the dataset
df = sns.load_dataset('tips')
df.head() 
```

#### 1.Create a Line Plot with x (‘total_bill’) versus y (‘tip’) using lineplot() function:

```python
sns.lineplot(x='total_bill',y='tip',data=df)
plt.xlabel("Total bill")
plt.ylabel("Tips")
plt.title("Tips given for Total Bill")
plt.show()
```

#### 2.Create a scatter plot with total_bill and tip 

```python
sns.scatterplot(x='total_bill',y='tip',data=df)
plt.xlabel("Total Bill")
plt.ylabel("Tips")
plt.show()
```

#### 3.Write a python code to create a hist plot on total_bill with bins=10.
- Create the following 
-Title as "Frequency of Bill Amount"
-xlabel as "Billed Amount"
-ylabel as "Count"

```python
sns.histplot(df.total_bill,bins=10)
plt.title("Frequency of Bill Amount")
plt.xlabel("Billed Amount")
plt.ylabel('Count')
plt.show()
```

## 4.Write a python code to create a heatmap on tips dataset by showing the corelation between numeric attributes 

```python
print(df.head(),'\n')
tips_corr = df.corr()
print("Correlation between numeric attributes\n",tips_corr,'\n')
sns.heatmap(tips_corr, annot=True, cmap='YlOrRd',annot_kws={'fontsize':11})
plt.title("Correlation of Numeric values in Tips Dataset")
plt.show()
```

#### 5.Write a python code to create a barplot on smoker and total_bill attribute

```python
sns.barplot(data=df,x='smoker',y='total_bill')
plt.xlabel("Smoker")
plt.ylabel("Total Bill ($)")
plt.show()
```

#### 6.Write a python code to use the countplot() function to show the count of smokers and non-smokers:

```python
sns.countplot(x=df.smoker,palette='prism_r')
plt.title("Count of Smokers and Non-Smokers")
plt.xlabel("Smoker")
plt.ylabel("Count")
plt.show()
```

#### 7.Let’s write a python code to analyze how the tipped amount varies during lunch and dinner time and if there are any outliers in our data. Hint :(Boxplot)

```python
sns.boxplot(x='time',y='tip',data=df)
plt.xlabel("TIME")
plt.ylabel("TIPS")
plt.title("Tipped Amount During Lunch And Dinner")
plt.show()
```

## The above plot has few outliers during Lunch time and many outliers during Dinner time

#### 8.Let’s write a python code to create a pair plot to show the pair wise distribution of all attributes

```python
sns.pairplot(data=df,hue='time',diag_kind='kde',kind='scatter',palette='flare',height = 2, aspect=2)
plt.show()
```
